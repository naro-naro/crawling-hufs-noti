# 영문, 빈 괄호 제거

```python
import re
```

```python
# 영문(알파벳 대문자, 소문자)을 제거하는 함수
def remove_english(text):
    return re.sub(r'[a-zA-Z]', '', text)

# 빈 괄호를 제거하는 함수
def remove_empty_parentheses(text):
    return re.sub(r'\(\)', '', text)
```

- 장학

```python
# index 1번째 글 후반부 'App'
janghak.head(2)
```

```python
# 영문 제거 함수 적용 - 장학
janghak['content'] = janghak['content'].apply(remove_english)
```

```python
# App 제거됨
janghak.head(2)
```

```python
# 빈 괄호 제거 함수 적용 - 장학
janghak['content'] = janghak['content'].apply(remove_empty_parentheses)
```

```python
# 빈 괄호 제거 확인
janghak.head(2)
```

# Komoran_key_sents

```python
import re
```

```python
from konlpy.tag import Komoran

# 초기화
komoran = Komoran()

# 명사, 동사, 형용사, 어간의 품사만 반환하는 함수
def komoran_tokenize(sent):
    words = komoran.pos(sent, join=True)
    words = [w for w in words if ('/NN' in w or '/XR' in w or '/VA' in w or '/VV' in w)]
    return words
```

```python
def komoran_key_sents(df, row):
  # df 데이터프레임의 row번째 행의 content를 문자열로 저장
  cont = df.loc[row, 'content']

  # 줄바꿈을 기준으로 문자열 분리하여 리스트에 저장
  sents = cont.split('\n')

  # 빈 문자열 제거
  sents = list(filter(None, sents))

  # 공백만 있는 문자열 제거
  sents = [sent for sent in sents if sent.strip()]

  # KeywordSummarizer 인스턴스 생성
  summarizer = KeysentenceSummarizer(tokenize = komoran_tokenize, min_sim = 0.3)

  # 핵심 3문장 요약 실행
  keysents = summarizer.summarize(sents, topk=3)

  # 반환
  return keysents[0][2], keysents[1][2], keysents[2][2]
```

# Okt_key_sents

```python
import re
```

```python
def okt_key_sents(df, row):
  # content가 빈 문자열이면 빈 리스트를 반환
  if df.loc[row, 'content'] == '':
      return []

  # df 데이터프레임의 row번째 행의 content를 문자열로 저장
  cont = df.loc[row, 'content']

  # 줄바꿈, 온점(.)을 기준으로 문자열 분리하여 리스트에 저장
  sents = re.split(r'[.\n]+', cont)
  # print(sents)

  # 3줄 미만 공지에 대비하여 추가
  sents.append('-')
  sents.append('-')

  # 빈 문자열 제거
  sents = list(filter(None, sents))
  # print(sents)

  # 공백만 있는 문자열 제거
  sents = [sent for sent in sents if sent.strip()]
  # print(sents)

  # 한 글자로 된 문자열이 포함되어 있는지 확인
  #   # 한 글자로 된 문자열을 리스트에서 삭제

  # 숫자로만 구성된 문자열이 포함되어 있는지 확인
  if any(element.isdigit() for element in sents):
    # 숫자로만 구성된 문자열을 리스트에서 삭제
    sents = [element for element in sents if not element.isdigit()]


  # 결과가 빈 리스트 일 때, 리스트를 그대로 반환****************
  if sents == []:
    return sents

  # 리스트 내의 요소가 2개 이하라면 리스트를 그대로 반환
  elif len(sents) <= 2:
    return sents

  elif sents.count('-') >= 1:
    sents = [s for s in sents if s != '-']
    return sents

  # KeywordSummarizer 인스턴스 생성
  summarizer = KeysentenceSummarizer(tokenize = okt_tokenize, min_sim = 0.3)

  # 핵심 3문장 요약 실행
  keysents = summarizer.summarize(sents, topk=3)

  # 반환
  return [keysents[0][2], keysents[1][2], keysents[2][2]]
```

# 오류 확인

```python
# 반복문을 사용하여 각 행에 대해 okt_key_sents 함수를 적용 - 장학
# 오류 : 33 -> 33 -> 22('-'부분 삭제, 빈리스트 반환) -> 13('-'부분 살리고, 빈리스트 반환) -> 7
error = 0
for i in range(len(janghak)):
    try:
        okt_key_sents(janghak, i)
    except Exception as e:
        print(f"오류 발생: {e} - 인덱스: {i}")
        error += 1
        continue
print(f'총 오류 수 : {error}')
```
